```python
# 导入 pandas 库并简称为 pd
import pandas as pd

# 创建一个字典 df1_data，用于存储数据
# 字典包含三个键值对，每个键代表一列数据
df1_data = {
    # '日期' 列存储日期信息，格式为 'YYYY-MM-DD'
    '日期': ['2021-08-21', '2021-08-22', '2021-08-23'],
    # '温度' 列存储对应的温度数值（单位：摄氏度）
    '温度': [25, 26, 50],
    # '湿度' 列存储对应的湿度百分比
    '湿度': [81, 50, 56]
}

# 使用 pd.DataFrame() 函数将字典转换为 DataFrame 对象
# 参数 data 指定要转换的数据，这里使用上面创建的 df1_data
df1 = pd.DataFrame(data=df1_data)

# 此时 df1 是一个 3 行 3 列的 DataFrame，包含日期、温度和湿度数据
# 索引默认为 0, 1, 2
```

```python
# 定义一个包含元组的列表作为数据源
# 每个元组代表一行数据，依次对应：日期、温度、湿度
df2_data = [
    ('2021-08-21', 25, 81),  # 第一行数据：2021-08-21的温度25、湿度81
    ('2021-08-22', 26, 50),  # 第二行数据：2021-08-22的温度26、湿度50
    ('2021-08-23', 27, 56)   # 第三行数据：2021-08-23的温度27、湿度56
]

# 使用pd.DataFrame()将列表转换为DataFrame
# 参数说明：
# - data：指定数据源，这里是上面定义的df2_data（元组列表）
# - columns：手动指定列名，顺序对应元组中数据的含义（日期、温度、湿度）
# - index：手动指定行索引，替代默认的0、1、2（这里用row_1、row_2、row_3）
df2 = pd.DataFrame(
    data=df2_data,
    columns=['日期', '温度', '湿度'],
    index=['row_1', 'row_2', 'row_3']
)

# 最终df2的结构：
#        日期  温度  湿度
# row_1  2021-08-21  25  81
# row_2  2021-08-22  26  50
# row_3  2021-08-23  27  56
df2
```

```python
import pandas as pd

# 创建一个示例数据集
data = {
    '性别': ['男', '女', '男', '女', '男', '女', '女', '男'],
    '购买': ['是', '否', '是', '是', '否', '否', '是', '否']
}

df = pd.DataFrame(data)

# 创建交叉表
crosstab = pd.crosstab(df['性别'], df['购买'])
print(crosstab)
```

```python
import pandas as pd

data = {
    '性别': ['男', '女', '男', '女', '男', '女'],
    '购买': ['是', '否', '是', '是', '否', '否'],
    '金额': [100, 150, 200, 130, 160, 120]
}
df = pd.DataFrame(data)

# 创建透视表
pivot_table = pd.pivot_table(df, values='金额', index='性别', columns='购买', aggfunc='mean')
print(pivot_table)
```
